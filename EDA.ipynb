### EDA - Exploratory Data Analysis (Analise exploratoria de dados)

```python
# importar as bibliotecas

import pandas as pd
import numpy as np
```

```python
# Ler a nossa base de dados

df = pd.read_csv('sales.csv',low_memory=False)
```

```python
pd.set_option('display.max_columns', None)
df.sample(5)
```

```python
df.groupby(['bi_st','status'])['total'].sum()
```

```python
df['Name Prefix'].unique()
```

```python
df.query('full_name == "Harper, Alvaro"')
```

```python
df.columns
```

```python
df.groupby('category')['total'].sum().reset_index().sort_values(by='total',ascending=False)
```

```python
df.sample(5)
```

```python
df.order_date.nunique()
```

```python
df.groupby('status')['order_id'].count().reset_index().sort_values(by='order_id',ascending=False)
```

```python
df.columns
```

```python
df.query('status == "canceled"')['payment_method'].value_counts()
```

```python
df.query('status == "complete"')['payment_method'].value_counts()
```

```python
(df.query('status == "canceled"').groupby(['category','sku'])['order_id']
 .count().reset_index().sort_values(by='order_id',ascending=False).head(10))
```

```python
df.query('sku == "MATSAM59DB75ADB2F80"')['status'].value_counts()
```

```python
df.groupby('State')['total'].sum().reset_index().sort_values(by='total',ascending=False)
```

```python
df.query('State == "TX"').groupby(['City'])['total'].sum().reset_index().sort_values(by='total',ascending=False)
```

```python
df.query('City == "Houston"').groupby(['category'])['total'].sum().reset_index().sort_values(by='total',ascending=False)
```
